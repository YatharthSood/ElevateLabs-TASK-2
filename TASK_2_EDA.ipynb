## VISUALISING DATA

```python
#HISTOGRAM

df['Age'].hist(bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
# BOXPLOT

sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
```

DIFFERENCE BETWEEN HISTOGRAM AND BOXPLOT
- Use histogram when you want to see the overall distribution shape (normal, skewed, multimodal).
- Use boxplot when you want to summarize data quickly and spot outliers or compare distributions across groups.
- In practice, use both together:
- Histogram → detailed distribution view.
- Boxplot → concise summary and anomaly detection.


## PAIRPLOT
- A pairplot is a grid of scatterplots and histograms.
- It shows relationships between multiple numeric features at once.

## CORRELATION MATRIX
- A correlation matrix is a table showing how strongly features are related.
- Each cell contains a correlation coefficient (between -1 and +1):
- +1 → perfect positive relationship (both increase together).
- -1 → perfect negative relationship (one increases, the other decreases).
- 0 → no linear relationship.



```python
# Pairplot
sns.pairplot(df[['Age','Fare','Survived']])
plt.show()

# Correlation matrix
corr = df.corr(numeric_only=True)
fig = px.imshow(corr, text_auto=True, title="Correlation Heatmap")
fig.show()
```
